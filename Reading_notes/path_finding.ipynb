```python
# ALL_WORDS is a set containing all strings that should be considered valid
# words (all in lower-case)
with open('words.txt') as f:
    ALL_WORDS = {i.strip() for i in f}

# replace the following with the starting state
start_state = "patties"

# replace this neighbors function:
def word_ladder_neighbors(state):
    """
    takes a state as input
    returns all neighboring states (valid words that differ in one letter)
    """

    legit = set()
    alphabets = 'abcdefghijklmnopqrstuvwxyz'
    for i in range(len(state)):
        for new_char in alphabets:
            new_word = state[:i] + new_char + state[i+1:]
            if new_word in ALL_WORDS:
                legit.add(new_word)
            else:
                continue
    return list(legit)

# replace this goal test function:
def goal_test_function(state):
    """
    takes a state as input
    returns True if and only if state matches the goal (the target word)
    """
    goal = 'foaming'
    if state == goal:
        return True
    else:
        return False

# ultimately, these variables will be passed as arguments to the find_path
# function to solve for the path between "patties" and "foaming"
# 



```

```python
def find_path(get_neighbours, start,goal_condition):
    if goal_condition(start):
        return (start,)
    agenda = [(start,)]
    visited = {start}
    while agenda:
        curr_path = agenda.pop(0)
        terminal_vertex = curr_path[-1]
        for child in get_neighbours(terminal_vertex):
            if child not in visited:
                new_path = curr_path + (child,)
                if goal_condition(child):
                    return new_path
                visited.add(child)
                agenda.append(new_path)
    return None
```

```python
def are_close_words(string1, string2):
    err_count = 0
    for s1, s2 in zip(string1, string2):
        if s1 != s2:
            err_count += 1
        else: 
            continue
    return True if err_count == 1 else False
```
